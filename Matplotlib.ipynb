### Matplotlib example workflow

```python
#0. import matplotlib (Already done at top) 

#1.  Prepare data
x=[1,2,3,4]
y=[10,40,50,90]

# 2. Setup Plot

fig, ax = plt.subplots(figsize=(5,5))

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title = "Simple plot",
       xlabel = "x-axis",
       ylabel = "y-axis")

# 5. Save & show (save the whole figure in your designated folder)
fig.savefig("images/sample-plot.png")
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond": 10,
                    "Peanut": 8,
                    "Cashew":12}
fig,ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax.set(title="Avneet's Nut Butter store",
      ylabel="Price ($)---->",
      xlabel="Butter Flavors ---->");
```

### Two options for subplots

```python
# Subplot option 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

#Plot to each different axis
ax1.plot(x,x/2);

ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
#Subplot option 2
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
# Plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

##### OO Method

* When plotting something quickly, use pyplot method
* When plotting something advanced and complex, use OO Method

```python
heart_disease.head()
```

```python
over50 = heart_disease[heart_disease["age"] > 50]
over50
```

```python
over50.plot(kind="scatter",
             x='age',
            y="chol",
            c='target');
```

```python
## OO Method

fix, a=plt.subplots(figsize=(10, 6))
over50.plot(kind='scatter',
           x='age',
           y='chol',
           c='target',
           ax=a);
a.set_xlim([45,100])
```

```python
## Oo method from scratch

fig, a = plt.subplots(figsize=(8,4))

#plot the data
scatter = a.scatter(x=over50["age"],
                     y=over50["chol"],
                     c=over50["target"],
                     );
a.set_xlim([50,80])

#Customize the plot
a.set(title="Heart Disease and Cholesterol Levels",
     xlabel="Age",
     ylabel="Cholestrol");
a.legend(*scatter.legend_elements(), title="Target"); #Add a legend

#Add a horizontal line
a.axhline(over50["chol"].mean(),
         linestyle=':');
```

```python
## Plotting over multiple axis
fig, (a0, a1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10)) 

#add data to a0
sc = a0.scatter(x=over50["age"],
               y=over50["chol"],
               c=over50["target"])
#Customize the plot
a0.set(title="Heart Disease and Cholesterol Levels",
     xlabel="Age",
     ylabel="Cholestrol");
a0.legend(*scatter.legend_elements(), title="Target"); #Add a legend

#Add a horizontal line
a0.axhline(over50["chol"].mean(),
         linestyle=':');

#add data to a1
sc = a1.scatter(x=over50["age"],
               y=over50["thalach"],
               c=over50["target"])
#Customize the plot
a1.set(title="Heart Disease and Thalach Levels",
     xlabel="Age",
     ylabel="Thalach");
a1.legend(*scatter.legend_elements(), title="Target"); #Add a legend

#Add a horizontal line
a1.axhline(over50["thalach"].mean(),
         linestyle=':');
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");

```

```python
#Set the style
plt.style.use("seaborn-whitegrid")

## Oo method from scratch

fig, a = plt.subplots(figsize=(10,6))

#plot the data
scatter = a.scatter(x=over50["age"],
                     y=over50["chol"],
                     c=over50["target"],
                     cmap="winter");
a.set_xlim([50,80])

#Customize the plot
a.set(title="Heart Disease and Cholesterol Levels",
     xlabel="Age",
     ylabel="Cholestrol");
a.legend(*scatter.legend_elements(), title="Target"); #Add a legend

#Add a horizontal line
a.axhline(over50["chol"].mean(),
         linestyle=':');
```

```python
# customizing y and x axis limitation

## Plotting over multiple axis
fig, (a0, a1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10)) 

#add data to a0
sc = a0.scatter(x=over50["age"],
               y=over50["chol"],
               c=over50["target"],
               cmap="winter")
#Customize the plot
a0.set(title="Heart Disease and Cholesterol Levels",
    
     ylabel="Cholestrol");
a0.set_xlim(50,80);
a0.legend(*scatter.legend_elements(), title="Target"); #Add a legend

#Add a horizontal line
a0.axhline(over50["chol"].mean(),
         linestyle='--');

#add data to a1
sc = a1.scatter(x=over50["age"],
               y=over50["thalach"],
               c=over50["target"],
               cmap="winter")
#Customize the plot
a1.set(title="Heart Disease and Thalach Levels",
     xlabel="Age",
     ylabel="Thalach");
a1.set_xlim(50,80);
a1.set_ylim(60,200);
a1.legend(*scatter.legend_elements(), title="Target"); #Add a legend

#Add a horizontal line
a1.axhline(over50["thalach"].mean(),
         linestyle='--');
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");

```
